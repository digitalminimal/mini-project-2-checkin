```python
def simplify(text):
    text=unicodedata.normalize('NFKD',text).encode('ascii','ignore').decode("utf-8") 
    spec_chars = ["!",'"',"#","%","&","'","(",")",
              "*","+",",","-",".","/",":",";","<",
              "=",">","?","@","[","\\","]","^","_",
              "`","{","|","}","~","–"]
    for char in spec_chars:
        text = text.replace(char, ' ')

    
    return(text)

simplify("434 Rue de l'Hôtel de Ville")
```
